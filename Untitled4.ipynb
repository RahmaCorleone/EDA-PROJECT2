```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd 
import numpy as np
```

```python
def file_path(path):
    file_name,file_extension=os.path.splitext(path)
    if file_extension=='.csv':
        df=pd.read_csv(path)
    elif file_extension=='.xlsx':
        df=pd.read_excel(path)
    else:
        raise ValueError("Unsupported file format. Please provide a CSV or XLS/XLSX file.")
    return df
data=file_path("International_Report_Passengers.csv")
data
```

```python
# identifying the data types of each column 
data.info()
```

```python
data.isna().sum()
```

```python
def preprocess(df):
    # Imputing missing values using median
    numerical_columns = df.select_dtypes(include=['number']).columns
    if not numerical_columns.empty:
        # If there are numerical columns
        df.fillna(data.median(numeric_only=True), inplace=True)
        
        # Normalize numerical columns using Min-Max Scaling (noramlization)
        for column in numerical_columns:
            df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())    

    return df.head()

data_processed = preprocess(data)
data_processed
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
def generate_histogram(column):
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x=column, bins=10)
    plt.title(f'Histogram of {column}')
    plt.show()

def generate_scatterplot(x_column, y_column):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=x_column, y=y_column)
    plt.title(f'Scatter plot of {x_column} vs {y_column}')
    plt.show()

data = file_path("International_Report_Passengers.csv")
data_processed = preprocess(data)

# Dropdown widgets for selecting x and y columns
x_column_selector = widgets.Dropdown(
    options=data.columns,
    description='Select x column:',
    disabled=False,
)

y_column_selector = widgets.Dropdown(
    options=data.columns,
    description='Select y column:',
    disabled=False,
)

# Button to trigger histogram generation
histogram_button = widgets.Button(description="Generate Histogram")

# Button to trigger scatter plot generation
scatterplot_button = widgets.Button(description="Generate Scatter Plot")

# Output areas for displaying the visualizations
histogram_output = widgets.Output()
scatterplot_output = widgets.Output()

# Function to generate histogram on button click
def on_histogram_button_clicked(b):
    histogram_output.clear_output()
    with histogram_output:
        generate_histogram(x_column_selector.value)

# Function to generate scatter plot on button click
def on_scatterplot_button_clicked(b):
    scatterplot_output.clear_output()
    with scatterplot_output:
        generate_scatterplot(x_column_selector.value, y_column_selector.value)

histogram_button.on_click(on_histogram_button_clicked)
scatterplot_button.on_click(on_scatterplot_button_clicked)

# Display widgets and output areas
display(x_column_selector, histogram_button, histogram_output)
display(y_column_selector, scatterplot_button, scatterplot_output)
```
